```python
# pip install opencv-python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

# 模糊化(Blur)

```python
smallBlur = np.ones((3, 3), dtype="float") * (1.0 / (3 * 3))
smallBlur
```

```python
# 使用cv2.filter2D進行convolution
image_name = './0_images/conv_org.png'
image = cv2.imread(image_name)
smallBlur = np.ones((7, 7), dtype="float") * (1.0 / (7 * 7))
result = cv2.filter2D(image, -1, smallBlur)
```

```python
# 顯示原圖與處理結果
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.imshow(image)
plt.subplot(122)
plt.imshow(result)
plt.show()
```

# 銳化(Sharpen)

```python
# construct a sharpening filter
sharpen = np.array((
	[0, -1, 0],
	[-1, 5, -1],
	[0, -1, 0]), dtype="int")

result = cv2.filter2D(image, -1, sharpen)
```

```python
# 顯示原圖與處理結果
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.imshow(image)
plt.subplot(122)
plt.imshow(result)
plt.show()
```

# 邊緣偵測(Edge detection)

```python
# construct the Laplacian kernel used to detect edge-like regions of an image
laplacian = np.array((
	[0, 1, 0],
	[1, -4, 1],
	[0, 1, 0]), dtype="int")

result = cv2.filter2D(image, -1, laplacian)
```

```python
# 顯示原圖與處理結果
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.imshow(image)
plt.subplot(122)
plt.imshow(result)
plt.show()
```

# X軸邊緣偵測

```python
# construct the Sobel x-axis kernel
sobelX = np.array((
	[-1, 0, 1],
	[-2, 0, 2],
	[-1, 0, 1]), dtype="int")

result = cv2.filter2D(image, -1, sobelX)
```

```python
# 顯示原圖與處理結果
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.imshow(image)
plt.subplot(122)
plt.imshow(result)
plt.show()
```

# Y軸邊緣偵測

```python
# construct the Sobel x-axis kernel
sobelY = np.array((
	[-1, -2, -1],
	[0, 0, 0],
	[1, 2, 1]), dtype="int")

result = cv2.filter2D(image, -1, sobelY)
```

```python
# 顯示原圖與處理結果
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.imshow(image)
plt.subplot(122)
plt.imshow(result)
plt.show()
```
